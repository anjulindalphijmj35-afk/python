```python
#leedcode no: 1
from collections import Counter
def findLucky(list):
    freq = Counter(list)
    lucky = -1
    for num, count in freq.items():
        if num == count:
            lucky = max(lucky, num)

    return lucky
print(findLucky([2,2,3,4]))
```

```python
def minSubArray(target, nums):
    left = 0
    sum = 0
    min_len = float('inf')

    for right in range(len(nums)):
        sum += nums[right]

        
        while sum >= target:
            min_len = min(min_len, right - left + 1)
            sum -= nums[left]
            left += 1
    return 0 if min_len == float('inf') else min_len      
```

```python
print(minSubArray (7,[2,3,1,2,4,3]))
```

```python
question set 3
def find_vowels(s):
    vowels = "aeiouAEIOU"
    result = []

    for ch in s:
        if ch in vowels:
            result.append(ch)
    
    return result
```

```python
s='RiFLuXYSS'
print(find_vowels(s))
```
