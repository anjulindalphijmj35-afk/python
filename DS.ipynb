```python
#iterating over keys
student={'a':10,'b':20}
for i in student.keys():
    print(i)
```

```python
##iterating ove values
student={'a':10,'b':20}
for i in student.values():
    print(i)
```

```python
#iterrate over key value pairs
student={'a':10,'b':20}
for i,value in student.items():
    print(i,value)
```

```python
#nested dictonaries
student={
    's1':{'name':'anand','age':32},
    's2':{'name':'kumar','age':35}
}
print(student)
```

```python
#access nested dictionaries elements student['s1']['name']
#iterating over nested dictionaries
for sid,info in student.items():
    print(sid)
    for k,v in info.items():
        print(k,v)
```

```python
#dictionary comphrehension
squares={x:x*2 for x in range(5)}
print(squares)
```

```python
#merge 2 dictonaries into one using
d1={'ss':1,'a':2}
d2={'s1':10,'s3':10}
print({**d1,**d2})
```

```python
#word count
n='hello welcome to trichy hello welcome to chennai'
words=n.split()
word_counts={word:words.count(word) for word in set (words)}
print(word_counts)
```

sets:
 .To store collections of unique items
 .They are unordered
 .They do not allow duplicate elements
 .Useful for membership tests,eliminating duplicate entries
 .Performing mathematical se opeations like union,intersection,difference and symmetric difference.

```python
#create a set {} set()
a={1,3,4,6}
type(a)
```

```python
#list of set
list(a)
```

```python
#adding elements
a.add(10)
print(a)
a.add(2)
print(a)
a.add('ravi')
print(a)
```

```python
#removing elements
a.remove(2)
print(a)
```

```python
#pop
a.pop()
print(a)
```

```python
#clear
a.clear()
print(a)
```

```python
#set membership test
my_set={1,2,3,4,5}
print(3 in my_set)
```

```python
#mathematical operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
```

```python
#union
set1.union(set2)
```

```python
#intersection
set1.intersection(set2)
```

```python
#differece
set2.difference(set1)
```

```python
#symmetric_difference
set1.symmetric_difference(set2)
```

```python
#sets words
set1={1,2,3,4,5}
set2={3,4}
```

```python
#issubset
set2.issubset(set1)
```

```python
#issuperset
set1.issuperset(set2)
```

```python
#counting unique words in text
text="hello welcome anand hello welcome ravi"
res=text.split(' ')
print(res)
des=set()
for i in res:
    des.add(i)
print(des)
```
