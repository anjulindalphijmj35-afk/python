```python
scope of variables
.local variable
.global variable
```

```python
# local variable
f=10
def fun():
    f='hi'
    print(f)
```

```python
fun()
print(f)
```

```python
f=10
def fun():
    global f
    f='hi'
    print(f)
```

```python
fun()
print(f)
```

```python
#using global variable
f=20
def fun():
    global f
    f='hi'
    print(f)
```

```python
fun()
print(f)
```

```python
 #recursion
def fact(n):
    if n==1:
        return n
    return n*fact(n-1)
```

```python
fact(5)
```

```python
# automation scripts or data fectching
def fetch_data():
    def retry():
        for i in range(3):
            print(f'attempt{i+1}')
            if i==0:
                return'success'
        return 'failed'
    return retry()
```

```python
print(fetch_data())
```

```python
def printonlyeven(n):
    for i in range(1,n+1):
        if i%2==0:
            print(i)
printonlyeven(10)
```

```python
def printonlyodd(n):
    for i in range(1,n+1):
        if i%2!=0:
            print(i)
printonlyodd(10)
```

```python
def printalleven(n,m):
    for i in range(n,m+1):
        if(i%2==0):
            print(i)
printalleven(10,20)
```

```python
def printallodd(n,m):
    for i in range(n,m+1):
        if(i%2!=0):
            print(i)
printallodd(11,20)
```

```python
def printalldivisable(n):
    for i in range(1,n):
        if(i%5==0):
            print(i)
printalldivisable(100)
```

```python
def issum(n):
    sum=0
    for i in range(1,n):
        sum=sum+1
        print(sum)
issum(10)
```
